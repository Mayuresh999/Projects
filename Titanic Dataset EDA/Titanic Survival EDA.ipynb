```python
def fun(c) : 
    age = c[0]
    pclass = c[1]
    
    if pd.isnull(age) : 
        if pclass == 1 : 
            return 39
        elif pclass == 2 : 
            return 34
        else : 
            return 27
    else : 
        return age
```
